```python
rating=[]
title=[]
link1=[]
price=[]
stock=[]
image_link=[]
book_link=[]
for i in tqdm(range(1,11)):
    link="https://books.toscrape.com/catalogue/page-"+str(i)+".html"
    v=requests.get(link)
    soup=BeautifulSoup(v.text,'html.parser')
    for sp in soup.find_all('li',class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):
        title.append(sp.find_all('a')[-1].get('title'))
        link1.append(sp.find_all('a')[-1].get('href'))
        rating.append(sp.find_all('p')[0].get('class')[1])
        price.append(sp.find_all('p')[1].text)
        stock.append(sp.find_all('p')[2].text.strip())
        book_link.append((sp.find_all('a')[0].get('href')))
        image_link.append((sp.find('img').get('src')))
    
    
```

```python
len(title)
```

```python
final={}
final['Book_Name']=title
final['Rating out of 5']=rating
final['Link']=link1
final['Price']=price
final['Quantity']=stock
final['Cover_Imagelink']=image_link
final['Book_Link']=book_link
```

```python
final
```

```python
df_f=pd.DataFrame(final)
```
